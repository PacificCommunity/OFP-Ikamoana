```python
import numpy as np
from ikamoana.ikamoanafields.ikamoanafields import IkamoanaFields
from ikamoana.ikamoanafields import sliceField
from ikamoana.feedinghabitat import coordsAccess
```

```python
file = './../data/feeding_habitat_data/po_2x30_historical_interim/skj_po_interim_2deg_configuration_file.xml'
ika_gen = IkamoanaFields(file)
```

# 1. Create Fields

```python
field_one = ika_gen.computeEvolvingTaxis(
    cohort_start=7, cohort_end=10,
    time_start=355, time_end=358,
    lat_min=7, lat_max=52,
    lon_min=15, lon_max=100)
```

```python
field_two = ika_gen.computeEvolvingTaxis(
    cohort_start=7,cohort_end=10,
    time_start=355,time_end=358,
    use_already_computed_habitat=False)
```

```python
A = field_one[0]
B = field_two[0]
print("A longitude :\n",A.lon.data[0],A.lon.data[-1]) 
print("\nB longitude :\n",B.lon.data[0],B.lon.data[-1])
```

# 2. Three way to resize
### a ) Using `reindex_like()`

```python
B.reindex_like(A)
```

### b ) Using `loc()`
> Will find the nearest values

```python
B.loc[min(A.time.data):max(A.time.data),
      max(A.lat.data):min(A.lat.data),
      min(A.lon.data):max(A.lon.data)]
```

```python
#min(A.time.data)
max(A.time.data)
```

### c ) Using `coordsAccess()` and `sliceField()`

```python
access_time, access_lat, access_lon = coordsAccess(B.coords)

minlon_idx = access_lon(min(A.lon.data))
maxlon_idx = access_lon(max(A.lon.data))
minlat_idx = access_lat(max(A.lat.data))
maxlat_idx = access_lat(min(A.lat.data))
mintime_idx = access_time(min(A.time.data))
maxtime_idx = access_time(max(A.time.data))
sliceField(B, mintime_idx,maxtime_idx, minlat_idx,maxlat_idx, minlon_idx,maxlon_idx)
```
